```python
df_tracks.set_index('release_date',inplace=True)
```

```python
df_tracks.index= pd.to_datetime(df_tracks.index)
df_tracks['year'] = pd.DatetimeIndex(df_tracks.index).year
df_tracks
```

```python
#convert duratyion from ms to sec
df_tracks['duration']=df_tracks['duration_ms'].apply(lambda x: round(x/1000))
```

```python
df_tracks.drop('duration_ms',inplace=True,axis=1)
```

```python
#which artist has the most monthly listeners for spotify
```

```python
corr_df=df_tracks.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
```

```python
corr_df
```

```python
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,cmap='inferno')
heatmap.set_title('correlation heatmap between variables')
#heatmap.set_xtick(heatnmap.get_xticklabls())
```

```python
#.4% of data
sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
```

```python
print(len(sample_df))
```

```python
#very high positive correlation 
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='loudness',x='energy',color='c').set(title='Loudness Vs Energy Correlation')
```

```python
#popn acousticness inversly related
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='popularity',x='acousticness',color='b').set(title='Popularity Vs acousticness')
```
